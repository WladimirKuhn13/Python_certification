```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(20)
```

```python
def getOneHotView(df):
  dfOneHotView = pd.DataFrame(columns = ['whoAmI_human', 'whoAmI_robot'])
  # dfTemporary
  humanTrue = {'whoAmI_human': True, 'whoAmI_robot': False}
  robotTrue = {'whoAmI_human': False, 'whoAmI_robot': True}

  dfHumanTrue = pd.DataFrame([humanTrue])
  dfRobotTrue = pd.DataFrame([robotTrue])

  for values in df['whoAmI']:
    if values == 'human':
      dfOneHotView.loc[len(dfOneHotView.index)] = humanTrue
    else:
      dfOneHotView.loc[len(dfOneHotView.index)] = robotTrue

  return dfOneHotView

df = getOneHotView(data)
df
```
